Клинические данные представляют собой  результаты лаботорных исследований, а также вычисленные значения, связанные с сердечно-сосудистым здоровьем:

Choi...25: Общий уровень холестерина (мг/дл)  
Tgi...26: Уровень триглицеридов (мг/дл)  
HDLi...27: Уровень "хорошего" ЛПВП-холестерина  
LDLi...28: Уровень "плохого" ЛПНП-холестерина (мг/дл)  
Choi...29: Соотношение общего холестерина/ЛПВП  
Tgi...30: Соотношение триглицеридов/ЛПВП  
HDLi...31: Уровень ЛПВП-холестерина (ммоль/л)  
LDLi...32: Уровень ЛПНП-холестерина (ммоль/л)  
AI: Коэффициент атерогенности (соотношение  ЛПНП/ЛПВП)  
r lat min: Минимальная толщина интима-медиального слоя правой латеральной сонной артерии (мм)  
r ant min: Минимальная толщина интима-медиального слоя правой передней сонной артерии (мм)  
r pos min: Минимальная толщина интима-медиального слоя правой задней сонной артерии (мм)  
l lat min: Минимальная толщина интима-медиального слоя левой латеральной сонной артерии (мм)  
l ant min: Минимальная толщина интима-медиального слоя левой передней сонной артерии (мм)  
l pos min: Минимальная толщина интима-медиального слоя левой задней сонной артерии (мм)  
r lat max: Максимальная толщина интима-медиального слоя правой латеральной сонной артерии (мм)  
r ant max: Максимальная толщина интима-медиального слоя правой передней сонной артерии (мм)  
r pos max: Максимальная толщина интима-медиального слоя правой задней сонной артерии (мм)  
l lat max: Максимальная толщина интима-медиального слоя левой латеральной сонной артерии (мм)  
l ant max: Максимальная толщина интима-медиального слоя левой передней сонной артерии (мм)  
l pos max: Максимальная толщина интима-медиального слоя левой задней сонной артерии (мм)  
carotid IMT mean: Средняя толщина интима-медиального слоя сонных артерий (мм)

Эти измерения важны для оценки сердечно-сосудистого здоровья пациента и факторов риска заболеваний сердца. Высокие уровни общего холестерина и "плохого" холестерина, а также низкие уровни "хорошего" холестерина связаны с повышенным риском заболеваний сердца. Утолщение интима-медиального слоя в сонных артериях также является фактором риска заболеваний сердца, поскольку оно может указывать на наличие атеросклероза, отложения бляшек в артериях.

**Переменные и шкалы измерений**

Возраст (Age) - количественная переменная, измеряемая в годах (шкала отношений).

Пол (Sex) - категориальная переменная, измеряемая номинально (мужской/женский).(номинальная шкала)

Индекс массы тела (BMI) - количественная переменная, измеряемая в кг/м2.(интервальная шкала)

Систолическое артериальное давление (Sys) - количественная переменная, измеряемая в мм рт. ст.(шкала отношений)

Диастолическое артериальное давление (Dia) - количественная переменная, измеряемая в мм рт. ст.(шкала отношений)

Группа (Group) - качественная переменная, измеряемая номинально (например, группа лечения или контрольная группа).(номинальная шкала)

 Choi...25, Tgi...26 - уровни, холестерина, триглицеридов и других веществ в крови, измеряемые  в мг/дл (миллиграмм на децилитр) (интервальная шкала)

### Гипотезы

Гипотеза: Статины снижают риск развития ишемической болезни сердца (IHD) у пациентов с высоким уровнем холестерина липопротеинов низкой плотности (LDL). Нулевая гипотеза: Статины не влияют на риск развития IHD у пациентов с высоким уровнем ЛПНП.
Альтернативная гипотеза: Статины снижают риск развития IHD у пациентов с высоким уровнем ЛПНП.

Гипотеза: У пациентов с диабетом (Dm) чаще встречается левожелудочковая гипертрофия (LVH), чем у пациентов без диабета. Нулевая гипотеза: Частота LVH у пациентов с диабетом не отличается от частоты LVH у пациентов без диабета.
Альтернативная гипотеза: Частота LVH у пациентов с диабетом выше, чем у пациентов без диабета.

Гипотеза: Увеличение индекса массы тела (BMI) у пациентов коррелирует с увеличением толщины интима-медиа сонной артерии (carotid IMT mean). Нулевая гипотеза: Индекс массы тела не коррелирует с толщиной интима-медиа сонной артерии у пациентов.
Альтернативная гипотеза: Индекс массы тела положительно коррелирует с толщиной интима-медиа сонной артерии у пациентов.

Гипотеза: Курение (Sm) увеличивает риск развития инсульта (Str) у пациентов. Нулевая гипотеза: Курение не влияет на риск развития инсульта у пациентов.
Альтернативная гипотеза: Курение увеличивает риск развития инсульта у пациентов.

Гипотеза: Пациенты с гипертонией (HT) имеют более низкий уровень холестерина липопротеинов высокой плотности (HDL), чем пациенты без гипертонии. Нулевая гипотеза: Уровень холестерина липопротеинов высокой плотности не отличается у пациентов с гипертонией и без гипертонии.
Альтернативная гипотеза: У пациентов с гипертонией уровень холестерина липопротеинов высокой плотности ниже, чем у пациентов без гипертонии.
